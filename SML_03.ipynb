```python
m=np.mean(d['latitude'])
print(m)
```

```python
v=np.var(d['latitude'])
print(v)
```

```python
n=len(d['latitude'])
print(n)
```

```python
MLE = (-n/2) * np.log(2 * np.pi) - (n/2) * np.log(v) - (1/(2 * v)) * (50 - m)**2
print("Log-likelihood value for 50 =", MLE)
```

```python
MLE = (-n/2) * np.log(2 * np.pi) - (n/2) * np.log(v) - (1/(2 * v)) * (75 - m)**2
print("Log-likelihood value for 75 =", MLE)
```

```python
MLE = (-n/2) * np.log(2 * np.pi) - (n/2) * np.log(v) - (1/(2 * v)) * (80 - m)**2
print("Log-likelihood value for 80 =", MLE)
```
