```python
col = ['Polyuria', 'Polydipsia', 'sudden weight loss', 'weakness', 'Polyphagia',
                      'Genital thrush', 'visual blurring', 'Itching', 'Irritability',
                      'delayed healing', 'partial paresis', 'muscle stiffness',
                      'Alopecia', 'Obesity']

d1[col] = d1[col].replace({'Yes': 1, 'No': 0})
d1['class'] = d1['class'].replace({'Positive': 1, 'Negative': 0})
d1['Gender'] = d1['Gender'].replace({'Male': 1, 'Female': 0})
d1
# here i have converted  the String "Yes" to 1 & "No" to 0.
# for class column "Positive" to 1 & "Negitive" to 0.
# for Gender column "Male" to 1 & "Female" to 0.
```

```python
d1.dtypes# checking the data types for verification
```

```python
y1=d1['class']   # Target
y1
```

```python
x1=d1.drop('class',axis=1) #features
x1
```

```python
# normalzing data
dd1=(x1-x1.min())/(x1.max()-x1.min())
dd1
```

```python
# graph b/w features and target
columns = ['Polyuria', 'Polydipsia', 'sudden weight loss', 'weakness', 'Polyphagia',
                      'Genital thrush', 'visual blurring', 'Itching', 'Irritability',
                      'delayed healing', 'partial paresis', 'muscle stiffness',
                      'Alopecia', 'Obesity']

for col in columns:
    x = d1[col]
    y = d1['class']
    plt.bar(x, y,edgecolor='k')
    plt.title(f'{col} vs class')
    plt.xlabel(f'Features  _ {col}')
    plt.ylabel('Target-class')
    plt.show()
```
